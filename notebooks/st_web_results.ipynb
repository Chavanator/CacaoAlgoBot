```python
import pandas as pd

# Cargar el archivo CSV
file_path = "D:\\iqrobot\\df_test\\operation_history.csv"
df = pd.read_csv(file_path)

# Convertir los resultados a 'win' y 'loss'
df['resultado'] = df['resultado'].map({85: 'win', 100: 'loss'})

# Calcular el número de operaciones ganadas y perdidas
wins = df[df['resultado'] == 'win'].shape[0]
losses = df[df['resultado'] == 'loss'].shape[0]

# Calcular el número de operaciones ganadas y perdidas por dirección
win_compras = df[(df['resultado'] == 'win') & (df['direccion'] == 'compra')].shape[0]
win_ventas = df[(df['resultado'] == 'win') & (df['direccion'] == 'venta')].shape[0]
loss_compras = df[(df['resultado'] == 'loss') & (df['direccion'] == 'compra')].shape[0]
loss_ventas = df[(df['resultado'] == 'loss') & (df['direccion'] == 'venta')].shape[0]

# Calcular rachas de operaciones ganadas y perdidas
streaks = df['resultado'].groupby((df['resultado'] != df['resultado'].shift()).cumsum()).apply(list)

win_streaks = [len(streak) for streak in streaks if streak[0] == 'win']
loss_streaks = [len(streak) for streak in streaks if streak[0] == 'loss']

max_win_streak = max(win_streaks, default=0)
max_loss_streak = max(loss_streaks, default=0)

# Contar rachas de ganancias y pérdidas
win_streak_count = {i: win_streaks.count(i) for i in range(1, max_win_streak + 1)}
loss_streak_count = {i: loss_streaks.count(i) for i in range(1, max_loss_streak + 1)}

# Resultados
print(f"Número de operaciones ganadas: {wins}")
print(f"Número de operaciones perdidas: {losses}")
print(f"Número de operaciones ganadas (compra): {win_compras}")
print(f"Número de operaciones ganadas (venta): {win_ventas}")
print(f"Número de operaciones perdidas (compra): {loss_compras}")
print(f"Número de operaciones perdidas (venta): {loss_ventas}")

print(f"Racha máxima de operaciones ganadas: {max_win_streak}")
print(f"Racha máxima de operaciones perdidas: {max_loss_streak}")

for i in range(1, max_win_streak + 1):
    print(f"Veces que se ganó {i} vez consecutiva: {win_streak_count[i]}")

for i in range(1, max_loss_streak + 1):
    print(f"Veces que se perdió {i} vez consecutiva: {loss_streak_count[i]}")

# Cálculos adicionales relevantes
total_operations = len(df)
win_percentage = (wins / total_operations) * 100 if total_operations > 0 else 0
loss_percentage = (losses / total_operations) * 100 if total_operations > 0 else 0

print(f"Porcentaje de operaciones ganadas: {win_percentage:.2f}%")
print(f"Porcentaje de operaciones perdidas: {loss_percentage:.2f}%")

# Probabilidad de ganar y perder por tipo de operación
total_compras = df[df['direccion'] == 'compra'].shape[0]
total_ventas = df[df['direccion'] == 'venta'].shape[0]

prob_win_compras = (win_compras / total_compras) * 100 if total_compras > 0 else 0
prob_win_ventas = (win_ventas / total_ventas) * 100 if total_ventas > 0 else 0
prob_loss_compras = (loss_compras / total_compras) * 100 if total_compras > 0 else 0
prob_loss_ventas = (loss_ventas / total_ventas) * 100 if total_ventas > 0 else 0

print(f"Probabilidad de ganar en compras: {prob_win_compras:.2f}%")
print(f"Probabilidad de ganar en ventas: {prob_win_ventas:.2f}%")
print(f"Probabilidad de perder en compras: {prob_loss_compras:.2f}%")
print(f"Probabilidad de perder en ventas: {prob_loss_ventas:.2f}%")
```
