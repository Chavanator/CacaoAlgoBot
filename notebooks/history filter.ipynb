```python
import pandas as pd

# Assuming your data is stored in a DataFrame called df
# Replace df with your actual DataFrame name if different

# Initialize variables
count_consecutive_8600_percent = 0
consecutive_count = 0

# Iterate over the 'P/G brutas' column
for value in df_filtered['Resultado']:
    if value == 'Ganada':
        consecutive_count += 1
        if consecutive_count == 2:
            count_consecutive_8600_percent += 1
            consecutive_count = 0
    else:
        consecutive_count = 0

# Print the count
print("Number of occurrences of 'Ganda' appearing four times in a row:", count_consecutive_8600_percent)
```

```python
# Initialize dictionary to store counts for different consecutive occurrences
consecutive_counts = {3: 0, 4: 0, 5: 0, 6:0, 7:0, 8:0, 9:0, 10:0}

# Initialize variable to keep track of consecutive occurrences
consecutive_count = 0

# Iterate over the 'P/G brutas' column
for value in df_filtered['Resultado']:
    if value == 'Ganada':
        consecutive_count += 1
        for num_consecutive in consecutive_counts:
            if consecutive_count == num_consecutive:
                consecutive_counts[num_consecutive] += 1
    else:
        consecutive_count = 0

# Print the counts for different consecutive occurrences
for num_consecutive, count in consecutive_counts.items():
    print(f"Number of occurrences of 'Ganada' appearing {num_consecutive} times consecutively:", count)
```
