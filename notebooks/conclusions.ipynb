```python
import os
import pandas as pd

# Ruta de la carpeta con los archivos CSV
carpeta = r'D:\iqrobot\resultados'

# Lista para almacenar cada DataFrame
dataframes = []

# Iterar sobre cada archivo en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):
        ruta_archivo = os.path.join(carpeta, archivo)
        df = pd.read_csv(ruta_archivo)
        dataframes.append(df)

# Combinar todos los DataFrames en uno solo
df_combined = pd.concat(dataframes, ignore_index=True)

# Extraer el par y la temporalidad del nombre del archivo
df_combined['Par'] = df_combined['Archivo'].apply(lambda x: x.split('_')[0])
df_combined['Temporalidad'] = df_combined['Archivo'].apply(lambda x: x.split('_')[1])

# Calcular la rentabilidad por par
rentabilidad_par = df_combined.groupby('Par')['Mínimo a ganar con el profit base'].mean().reset_index()
par_mas_rentable = rentabilidad_par.loc[rentabilidad_par['Mínimo a ganar con el profit base'].idxmax()]

# Calcular la rentabilidad por temporalidad
rentabilidad_temporalidad = df_combined.groupby('Temporalidad')['Mínimo a ganar con el profit base'].mean().reset_index()
temporalidad_mas_rentable = rentabilidad_temporalidad.loc[rentabilidad_temporalidad['Mínimo a ganar con el profit base'].idxmax()]

# Calcular la rentabilidad por ciclo de interés compuesto
columnas_interes = [
    'Interés compuesto a un 85% para 2 ciclos',
    'Interés compuesto a un 85% para 3 ciclos',
    'Interés compuesto a un 85% para 4 ciclos'
]
rentabilidad_ciclo = df_combined[columnas_interes].mean().reset_index()
ciclo_mas_rentable = rentabilidad_ciclo.loc[rentabilidad_ciclo[0].idxmax()]

# Mostrar los resultados
print(f"Par más rentable: {par_mas_rentable['Par']} con una rentabilidad media de {par_mas_rentable['Mínimo a ganar con el profit base']}")
print(f"Temporalidad más rentable: {temporalidad_mas_rentable['Temporalidad']} con una rentabilidad media de {temporalidad_mas_rentable['Mínimo a ganar con el profit base']}")
print(f"Ciclo de interés compuesto más rentable: {ciclo_mas_rentable['index']} con una rentabilidad media de {ciclo_mas_rentable[0]}")

# Conclusión general
print("\nConclusión general:")
print(f"El par más rentable es {par_mas_rentable['Par']}.")
print(f"La temporalidad más rentable es {temporalidad_mas_rentable['Temporalidad']}.")
print(f"El ciclo de interés compuesto más rentable es {ciclo_mas_rentable['index']}.")

# Guardar el DataFrame combinado en un nuevo archivo CSV
df_combined.to_csv(r'D:\iqrobot\resultados_combined.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('D:\\iqrobot\\resultados_combined2.csv')

# Asegurarnos de que las columnas relevantes estén en formato numérico
for col in [
    'Número de resultados 85', 'Número de -100', 'Número de veces que aparece 85 2 veces consecutivas',
    'Número de veces que aparece 85 3 veces consecutivas', 'Número de veces que aparece 85 4 veces consecutivas',
    'Número de veces que aparece -100 2 veces consecutivas', 'Número de veces que aparece -100 3 veces consecutivas',
    'Número de veces que aparece -100 4 veces consecutivas', 'Total de pérdida', 'Mínimo a ganar con el profit base',
    'profit1', 'Interés compuesto a un 85% para 2 ciclos', 'profit2', 'Interés compuesto a un 85% para 3 ciclos',
    'profit3', 'Interés compuesto a un 85% para 4 ciclos', 'profit4'
]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Filtrar las columnas numéricas para calcular la matriz de correlación
df_numeric = df.select_dtypes(include=[float, int])

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Mostrar la correlación con respecto a 'Mínimo a ganar con el profit base'
print("Correlación con respecto a 'Mínimo a ganar con el profit base':")
print(correlation_matrix['Mínimo a ganar con el profit base'].sort_values(ascending=False))

# Visualizar la matriz de correlación
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
