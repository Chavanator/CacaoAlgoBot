```python
import os
import time
import pandas as pd
from datetime import datetime
from tvDatafeed import TvDatafeed, Interval

# Configuración de TradingView
tv = TvDatafeed()
tv.__ws_timeout = 60  # Aumentamos el tiempo de espera para evitar timeouts

# Parámetros de descarga
symbol = "BTCUSDT"
exchange = "BINANCE"
n_datasets = 1000  # Número de datasets por temporalidad
n_bars = 506  # Cantidad de velas por dataset
save_dir = r"D:\YOP\iqoptionapi-master\Prueba_df"

# Crear la carpeta si no existe
os.makedirs(save_dir, exist_ok=True)

# Configuración de intervalos
intervals = {
    "4h": Interval.in_4_hour,
    "1h": Interval.in_1_hour,
    "15m": Interval.in_15_minute
}

def get_data_with_retry(symbol, exchange, interval, n_bars, retries=3):
    """ Intenta descargar datos de TradingView con reintentos en caso de fallo. """
    for i in range(retries):
        try:
            data = tv.get_hist(symbol=symbol, exchange=exchange, interval=interval, n_bars=n_bars)
            if data is not None:
                return data
        except Exception as e:
            print(f"Intento {i+1} fallido: {e}")
            time.sleep(5)  # Esperar 5 segundos antes de reintentar
    return None  # Si falla después de los intentos, devuelve None

# Descargar datos para cada temporalidad
for timeframe, interval in intervals.items():
    print(f"\nDescargando {n_datasets} datasets para {timeframe}...")

    for i in range(n_datasets):
        print(f"Descargando dataset {i+1}/{n_datasets} para {timeframe}...")

        # Obtener datos
        data = get_data_with_retry(symbol, exchange, interval, n_bars)
        
        if data is None:
            print(f"⚠️ No se pudieron obtener datos para {timeframe} en el dataset {i+1}.")
            continue  # Saltar al siguiente dataset
        
        # Convertir a DataFrame y resetear el índice
        df = pd.DataFrame(data)
        df.reset_index(inplace=True)
        
        # Guardar como CSV
        file_path = os.path.join(save_dir, f"{symbol}_{timeframe}_dataset_{i+1}.csv")
        df.to_csv(file_path, index=False)
        print(f"✅ Datos guardados en {file_path}")

print("\n🚀 ¡Descarga completada con éxito!")
```
